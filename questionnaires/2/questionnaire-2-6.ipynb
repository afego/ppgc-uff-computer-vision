```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("q6.jpg")
plt.imshow(img)
plt.show()
```

```python
intensity_levels = [128,64,32]

for intensity_level in intensity_levels:
    step = 256//intensity_level
    bins = [x for x in range(0,256,step)]
    bin_size = len(bins)
    new_img = np.zeros_like(img)
    for i in range(1,bin_size):
        new_img[(img >= bins[i-1]) & (img < bins[i])] = bins[i-1]
    new_img[(img > bins[-1])] = 255
    
    cv.imwrite(f"intensity-{intensity_level}.png", new_img)
```
