Apply the median filter on the "Questionario-3-Imagem-3.tif" image to reduce impulsive salt and pepper noise. Vary the kernel size between 3x3, 5x5 and 7x7 and observe the results

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
filename = "Questionario-3-Imagem-3"
full_name = filename+".tif"

img = cv.imread(full_name)
```

```python
fig = plt.figure(figsize=(15,15))
for i in range(1, 9, 2):
    ax = fig.add_subplot(5,3,i)
    blur_img = cv.medianBlur(img, i)
    ax.title.set_text(f"Kernel {i}x{i}")
    plt.imshow(blur_img)
    plt.imsave(f"{filename}_{i}x{i}.png",blur_img)
plt.show
```

Conforme o kernel aumentou, a quantidade de informaes perdidas tambm aumentou. Isso ocorre pois, com o aumento da vizinhana, maior  a quantidade de valores prximos utilizados na convoluo, removendo os valores mais discrepantes na regio.
