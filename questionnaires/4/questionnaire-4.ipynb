```python
import numpy as np
import cv2 as cv
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
img1 = cv.imread("Questionario-4-Bricks1.jpg")
img2 = cv.imread("Questionario-4-Bricks2.jpg")

# img1 = cv.imread("Questionario-4-Building1.jpg")
# img2 = cv.imread("Questionario-4-Building2.jpg")
```

SIFT

```python
sift = cv.SIFT_create()

keyp1, descr1 = sift.detectAndCompute(img1, None)
keyp2, descr2 = sift.detectAndCompute(img2, None)

bfm = cv.BFMatcher()
matches = bfm.knnMatch(descr1, descr2, k=2)

ratio_test = []
for m,n in matches:
    if m.distance < 0.8*n.distance:
        ratio_test.append([m])

img_match = cv.drawMatchesKnn(img1, keyp1, img2, keyp2, ratio_test, None, flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(img_match)
plt.show()
```
